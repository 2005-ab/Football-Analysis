```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the image
image_path = r"D:\ab\strategy\output_videos\cropped_image.jpg"
image = cv2.imread(image_path)

if image is None:
    raise FileNotFoundError(f"Image not found at: {image_path}")

# Convert from BGR to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Get the top half of the image
top_half_image = image[:image.shape[0] // 2, :, :]

# Flatten the top half to 2D for clustering
image_2d = top_half_image.reshape(-1, 3)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(image_2d)
labels = kmeans.labels_

# Reshape the labels back to 2D image format
clustered_image = labels.reshape(top_half_image.shape[0], top_half_image.shape[1])

# Display the clustered image
plt.imshow(clustered_image, cmap='viridis')
plt.title("KMeans Clustering on Top Half")
plt.axis('off')
plt.show()
```

```python
corner_clusters =[clustered_image[0,0], clustered_image[0,-1], clustered_image[-1,0], clustered_image[-1,-1]]
non_player_cluster= max(set(corner_clusters), key=corner_clusters.count)
print(non_player_cluster)
```

```python
player_cluster=1-non_player_cluster
print(player_cluster)
```

```python
player_rgb = kmeans.cluster_centers_[player_cluster]
print("RGB value for player cluster:", player_rgb)
```
